```python
import spacy

# Load the English NLP model
nlp = spacy.load("en_core_web_sm")  

# Or load the medical model:
# nlp_medical = spacy.load("en_core_sci_md")

# Sample text
text = "Dr. John Doe, a cardiologist at Mayo Clinic, prescribed Aspirin 100mg."

# Process the text
doc = nlp(text)  # Use nlp_medical for medical text
```

```python
for token in doc:
    print(token.text)
```

```python
for ent in doc.ents:
    print(ent.text, " -> ", ent.label_)
```

```python
for token in doc:
    print(token.text, " -> ", token.pos_)
```

```python
for token in doc:
    print(token.text, " -> ", token.dep_, " -> ", token.head.text)
```

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
pattern = [{"LOWER": "aspirin"}, {"IS_DIGIT": True}, {"LOWER": "mg"}]
matcher.add("MEDICATION_PATTERN", [pattern])

matches = matcher(doc)
for match_id, start, end in matches:
    print(doc[start:end].text)
```

```python
word1 = nlp("doctor")
word2 = nlp("nurse")
print(word1.similarity(word2))
```
